## MatEnsemble Demo

First we import the **Pipline** object from matensemble and create an instance of one

```python
from matensemble.pipeline import Pipeline 
from mpi4py import MPI

pipe = Pipeline()
```

Next we create some chores that we want MatEnsemble schedule in our workflow

```python
for _ in range(10):
    pipe.exec(command=["echo", "Hello World!"]) # Simple echo hello world jobs
```

Now that we have created all of the 'Chores' we can submit the workflow

```python
pipe.submit()
```

MatEnsemble will then create the DAG, topologically sort it, and submit that 
list to the manager. The manager will connect to the flux broker and begin 
scheduling the chores.
