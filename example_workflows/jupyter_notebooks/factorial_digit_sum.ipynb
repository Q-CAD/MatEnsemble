```python
from matensemble.pipeline import Pipeline

pipe = Pipeline()
```

```python
@pipe.chore()
def factorial(n: int): 
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

@pipe.chore()
def digit_sum(n: int): 
    sum = 0
    for char in str(n):
        sum += int(char)
    return sum
```

```python
a = factorial(100)
b = digit_sum(a)
```

```python
pipe.submit()
print(f"The factorial digit sum of 100 is {b}")
```
