```python
from matensemble.pipeline import Pipeline

pipe = Pipeline()
```

```python
@pipe.chore()
def factorial(n: int): 
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

@pipe.chore()
def digit_sum(n: int): 
    sum = 0
    for char in str(n):
        sum += int(char)
    return sum
```

```python
results = []
for i in range(10):
    a = factorial(i)
    b = digit_sum(a)
    results.append(b)
```

```python
pipe.submit()
for result in results:
    print(result)
```
